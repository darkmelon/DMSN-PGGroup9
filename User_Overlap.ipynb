```python
import pandas as pd

bitcoin = pd.read_csv('bitcoin.csv')
crypto = pd.read_csv('crypto.csv')
stocks = pd.read_csv('stocks.csv')
wsb = pd.read_csv('wsb.csv')
```

```python
bitcoin.head()
```

#BTC - Crypto Overlap

```python
btc_cryp = []

for i in bitcoin["Id"]:
  for j in crypto["Id"]:
    if i in j:
      btc_cryp.append(bitcoin[bitcoin["Id"]==i].values)
      btc_cryp.append(crypto[crypto["Id"]==j].values)

btc_cryp    
```

```python
btc_cryp_pd = []

for i in btc_cryp:
  btc_cryp_pd.append(i[0])

btc_cryp_pd
```

```python
# dataframe showing stats for users that appear in bitcoin and crypto repsectively
pd.set_option('display.max_rows', 1000)
df_btc_cryp = pd.DataFrame(btc_cryp_pd, columns=list(bitcoin.columns))
df_btc_cryp
```

#BTC - WSB Overlap

```python
btc_wsb = []

for i in bitcoin["Id"]:
  for j in wsb["Id"]:
    if i in j:
      btc_wsb.append(bitcoin[bitcoin["Id"]==i].values[0])
      btc_wsb.append(wsb[wsb["Id"]==j].values[0][:15])

btc_wsb          
```

```python
df_btc_wsb = pd.DataFrame(btc_wsb, columns=list(bitcoin.columns))
df_btc_wsb

```

#BTC-Crypto-Stocks-WSB Overlap

```python
btc_cryp_stocks_wsb = []

for i in bitcoin["Id"]:
  for j in crypto["Id"]:
    if i in j:
      btc_cryp_stocks_wsb.append(bitcoin[bitcoin["Id"]==i].values[0])
      btc_cryp_stocks_wsb.append(crypto[crypto["Id"]==j].values[0][:15])
    for k in stocks["Id"]:
      if i in k:
        btc_cryp_stocks_wsb.append(stocks[stocks["Id"]==k].values[0][:15])
      for l in wsb["Id"]:
        if i in l:
          btc_cryp_stocks_wsb.append(wsb[wsb["Id"]==l].values[0][:15])

btc_cryp_stocks_wsb
```

```python
df_btc_cryp_stocks_wsb = pd.DataFrame(btc_cryp_stocks_wsb, columns=list(bitcoin.columns))
df_btc_cryp_stocks_wsb
```

#Stocks-WSB Overlap

```python
stocks_wsb = []

for i in stocks["Id"]:
  for j in wsb["Id"]:
    if i in j:
      stocks_wsb.append(stocks[stocks["Id"]==i].values[0])
      stocks_wsb.append(wsb[wsb["Id"]==j].values[0])
```

```python
df_stocks_wsb = pd.DataFrame(stocks_wsb, columns=list(stocks.columns))
df_stocks_wsb
```
